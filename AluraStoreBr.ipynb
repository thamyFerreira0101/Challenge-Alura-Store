#1. Análise do faturamento


```python
import pandas as pd

# Exemplo de dados para cada loja
data_loja1 = {
    "Produto": ["Assistente virtual", "Mesa de jantar"],
    "Categoria do Produto": ["eletronicos", "moveis"],
    "Preço": [219.08, 256.35],
    "Frete": [9.25, 11.23]
}

data_loja2 = {
    "Produto": ["Jogo de tabuleiro", "Micro-ondas"],
    "Categoria do Produto": ["brinquedos", "eletrodomesticos"],
    "Preço": [279.51, 1009.99],
    "Frete": [21.26, 54.67]
}

data_loja3 = {
    "Produto": ["Cadeira de escritório"],
    "Categoria do Produto": ["moveis"],
    "Preço": [446.99],
    "Frete": [26.96]
}

data_loja4 = {
    "Produto": ["Televisão", "Sofá"],
    "Categoria do Produto": ["eletronicos", "moveis"],
    "Preço": [1500.00, 800.00],
    "Frete": [50.00, 30.00]
}

# Criar DataFrames para cada loja
loja1 = pd.DataFrame(data_loja1)
loja2 = pd.DataFrame(data_loja2)
loja3 = pd.DataFrame(data_loja3)
loja4 = pd.DataFrame(data_loja4)

# Concatenar os dados das quatro lojas
df_total = pd.concat([loja1, loja2, loja3, loja4])

# Calcular o faturamento total
faturamento_total = df_total["Preço"].sum() + df_total["Frete"].sum()

# Calcular o faturamento por categoria
df_total['Faturamento'] = df_total['Preço'] + df_total['Frete']
faturamento_por_categoria = df_total.groupby('Categoria do Produto')['Faturamento'].sum().reset_index()

# Exibir resultados
print(f"Faturamento total: R$ {faturamento_total:,.2f}")
print("\nFaturamento por Categoria:")
print(faturamento_por_categoria)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Exemplo de dados para cada loja
data_loja1 = {
    "Produto": ["Assistente virtual", "Mesa de jantar"],
    "Categoria do Produto": ["eletronicos", "moveis"],
    "Preço": [219.08, 256.35],
    "Frete": [9.25, 11.23],
    "Loja": ["Loja 1", "Loja 1"]  # Adicionando a coluna de loja
}

data_loja2 = {
    "Produto": ["Jogo de tabuleiro", "Micro-ondas"],
    "Categoria do Produto": ["brinquedos", "eletrodomesticos"],
    "Preço": [279.51, 1009.99],
    "Frete": [21.26, 54.67],
    "Loja": ["Loja 2", "Loja 2"]  # Adicionando a coluna de loja
}

data_loja3 = {
    "Produto": ["Cadeira de escritório"],
    "Categoria do Produto": ["moveis"],
    "Preço": [446.99],
    "Frete": [26.96],
    "Loja": ["Loja 3"]  # Adicionando a coluna de loja
}

data_loja4 = {
    "Produto": ["Televisão", "Sofá"],
    "Categoria do Produto": ["eletronicos", "moveis"],
    "Preço": [1500.00, 800.00],
    "Frete": [50.00, 30.00],
    "Loja": ["Loja 4", "Loja 4"]  # Adicionando a coluna de loja
}

# Criar DataFrames para cada loja
loja1 = pd.DataFrame(data_loja1)
loja2 = pd.DataFrame(data_loja2)
loja3 = pd.DataFrame(data_loja3)
loja4 = pd.DataFrame(data_loja4)

# Concatenar os dados das quatro lojas
df_total = pd.concat([loja1, loja2, loja3, loja4])

# Calcular faturamento (Preço + Frete)
df_total['Faturamento'] = df_total['Preço'] + df_total['Frete']

# Calcular faturamento total por loja
faturamento_por_loja = df_total.groupby('Loja')['Faturamento'].sum().reset_index()

# Mostrar faturamento total no console
print(faturamento_por_loja)

# Criar gráfico de barras para faturamento por loja
plt.figure(figsize=(10,6))
bars = plt.bar(faturamento_por_loja['Loja'], faturamento_por_loja['Faturamento'], color='skyblue')
plt.title('Faturamento por Loja', fontsize=16)
plt.xlabel('Loja', fontsize=14)
plt.ylabel('Faturamento (R$)', fontsize=14)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adicionando os valores em cima das barras
for bar in bars:
    height = bar.get_height()
    plt.annotate(f'R$ {height:,.2f}',
                 xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 5),  # 5 points vertical offset
                 textcoords="offset points",
                 ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

# 3. Média de Avaliação das Lojas

```python


# Exemplo de dados para cada loja com as colunas de interesse
data_loja1 = {
    "Produto": ["Assistente virtual", "Mesa de jantar"],
    "Categoria do Produto": ["eletronicos", "moveis"],
    "Preço": [219.08, 256.35],
    "Frete": [9.25, 11.23],
    "Loja": ["Loja 1", "Loja 1"],
    "Tipo de pagamento": ["cartao_credito", "cartao_credito"],
    "Quantidade de parcelas": [8, 4]
}

data_loja2 = {
    "Produto": ["Jogo de tabuleiro", "Micro-ondas"],
    "Categoria do Produto": ["brinquedos", "eletrodomesticos"],
    "Preço": [279.51, 1009.99],
    "Frete": [21.26, 54.67],
    "Loja": ["Loja 2", "Loja 2"],
    "Tipo de pagamento": ["cartao_credito", "boleto"],
    "Quantidade de parcelas": [1, 1]
}

data_loja3 = {
    "Produto": ["Cadeira de escritório"],
    "Categoria do Produto": ["moveis"],
    "Preço": [446.99],
    "Frete": [26.96],
    "Loja": ["Loja 3"],
    "Tipo de pagamento": ["boleto"],
    "Quantidade de parcelas": [1]
}

data_loja4 = {
    "Produto": ["Televisão", "Sofá"],
    "Categoria do Produto": ["eletronicos", "moveis"],
    "Preço": [1500.00, 800.00],
    "Frete": [50.00, 30.00],
    "Loja": ["Loja 4", "Loja 4"],
    "Tipo de pagamento": ["cartao_credito", "cartao_credito"],
    "Quantidade de parcelas": [6, 5]
}

# Criar DataFrames para cada loja
loja1 = pd.DataFrame(data_loja1)
loja2 = pd.DataFrame(data_loja2)
loja3 = pd.DataFrame(data_loja3)
loja4 = pd.DataFrame(data_loja4)

# Concatenar os dados das quatro lojas
df_total = pd.concat([loja1, loja2, loja3, loja4], ignore_index=True)

# Calcular faturamento (Preço + Frete)
df_total['Faturamento'] = df_total['Preço'] + df_total['Frete']

# Criar coluna indicando se foi parcelado ou não (sim se parcelas > 1)
df_total['Parcelado'] = df_total['Quantidade de parcelas'].apply(lambda x: 'Sim' if x > 1 else 'Não')

# Agrupar por loja, tipo de pagamento e parcelamento, calculando média de faturamento
media_faturamento = df_total.groupby(
    ['Loja', 'Tipo de pagamento', 'Parcelado', 'Quantidade de parcelas']
)['Faturamento'].mean().reset_index()

# Renomear a coluna para melhor entendimento
media_faturamento.rename(columns={'Faturamento': 'Média de Faturamento (R$)'}, inplace=True)

# Exibir o resultado
print(media_faturamento)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configurar o estilo do gráfico
sns.set(style="whitegrid")

# Criar o gráfico de barras
plt.figure(figsize=(12, 6))
bar_plot = sns.barplot(
    data=media_faturamento,
    x='Loja',
    y='Média de Faturamento (R$)',
    hue='Parcelado',
    palette='viridis'
)

# Adicionar título e rótulos
plt.title('Média de Faturamento por Loja e Tipo de Pagamento')
plt.xlabel('Loja')
plt.ylabel('Média de Faturamento (R$)')
plt.legend(title='Parcelado')

# Exibir o gráfico
plt.show()
```

# 5. Frete Médio por Loja


```python
# Dados dos produtos
dados = [
    {"Produto": "Assistente virtual", "Frete": 9.249790},
    {"Produto": "Mesa de jantar", "Frete": 11.234305},
    {"Produto": "Jogo de tabuleiro", "Frete": 21.262681},
    {"Produto": "Micro-ondas", "Frete": 54.667344},
    {"Produto": "Cadeira de escritório", "Frete": 26.964689},
]

# Função para calcular a média do frete
def calcular_media_frete(dados):
    total_frete = 0
    quantidade_produtos = len(dados)

    for produto in dados:
        total_frete += produto["Frete"]

    media_frete = total_frete / quantidade_produtos
    return media_frete

# Chamando a função e imprimindo o resultado
media = calcular_media_frete(dados)
print(f"A média do frete dos produtos é: {media:.2f}")

```
