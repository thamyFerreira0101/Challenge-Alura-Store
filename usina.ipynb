```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('usina.csv')
```

```python
df.head()
```

```python
y = df['PE']
X = df.drop(columns=['PE'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=230)
```

```python
import statsmodels.api as sm
```

```python
#adicionando o constante
X_train= sm.add_constant(X_train)
```

```python
modelo = sm.OLS(y_train,X_train).fit()
```

```python
modelo.summary()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
X_test.columns
```

```python
explicativas_1 = ['AT', 'V', 'AP', 'RH']
```

```python
# VIF 1
vif_1 = pd.DataFrame()
vif_1["variavel"] = explicativas_1

vif_1["vif"] = [variance_inflation_factor(X_train[explicativas_1].values, i)
                          for i in range(len(explicativas_1))]

print(vif_1)
```

```python
import plotly.express as px
```

```python
Y_previsto_train =  modelo.predict(X_train)
```

```python
fig = px.scatter(x=Y_previsto_train, y=y_train,
                 title='Previsão X Real',
                 labels={'x':'Preço previsto','y':'Preço real'})
fig.show()
```

```python
# Residuos
residuos = modelo.resid
```

```python
fig = px.scatter(x=Y_previsto_train, y=residuos,
                 title='Previsão X Real',
                 labels={'x':'Preço previsto','y':'Resíduos'})
fig.show()
```
